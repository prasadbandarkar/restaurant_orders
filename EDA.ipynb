```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rest1_orders=pd.read_csv('restaurant-1-orders.csv',parse_dates=['Order Date'],dayfirst=True)
#rest1_orders.columns=['order_number','date','item_name','quantity','product_price','total_products']
rest1_prices=pd.read_csv('restaurant-1-products-price.csv')
#rest1_prices.columns=['item_name','product_price']
```

```python
order_price=rest1_orders.quantity.mul(rest1_orders.product_price).groupby(rest1_orders.order_number).sum().reset_index(name='total_price')
rest1_orders=rest1_orders.merge(order_price,how='left',on='order_number')
```
